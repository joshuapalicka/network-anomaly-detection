```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from scipy.io.arff import loadarff
import pandas as pd
import numpy as np
```

```python
dataframe = pd.DataFrame(loadarff("nsl-kdd/KDDTrain+.arff")[0])
dataframe
```

```python
test_dataframe = pd.DataFrame(loadarff("nsl-kdd/KDDTest+.arff")[0])
test_dataframe
```

```python
dataframe = dataframe.astype({'protocol_type':'category', 
                              'service':'category', 
                              'flag':'category', 
                              'land':bool, 
                              'logged_in':bool, 
                              'is_host_login':bool, 
                              'is_guest_login':bool})
test_dataframe = test_dataframe.astype({'protocol_type':'category', 
                                        'service':'category', 
                                        'flag':'category', 
                                        'land':bool, 
                                        'logged_in':bool, 
                                        'is_host_login':bool, 
                                        'is_guest_login':bool})
dataframe.dtypes
```

```python
dataframe['service'] = dataframe['service'] == b'tcp'
test_dataframe['service'] = test_dataframe['service'] == b'tcp'

features = ['duration', 'service', 'src_bytes', 'dst_bytes', 'land', 'wrong_fragment', 'urgent', 'hot', 'num_failed_logins', 'logged_in', 'num_compromised',
            'root_shell', 'su_attempted', 'num_root', 'num_file_creations', 'num_shells', 'num_access_files', 'num_outbound_cmds', 'is_host_login',
            'is_guest_login', 'count', 'srv_count', 'serror_rate', 'srv_serror_rate', 'rerror_rate', 'srv_rerror_rate', 'same_srv_rate',
            'diff_srv_rate', 'srv_diff_host_rate', 'dst_host_count', 'dst_host_srv_count', 'dst_host_same_srv_rate', 'dst_host_diff_srv_rate',
            'dst_host_same_src_port_rate', 'dst_host_srv_diff_host_rate', 'dst_host_serror_rate', 'dst_host_srv_serror_rate', 'dst_host_rerror_rate',
            'dst_host_srv_rerror_rate']
#categorical strong features: 'flag'
strong_features = ['service', 'dst_bytes', 'count', 'serror_rate', 'srv_serror_rate', 'srv_rerror_rate', 'same_srv_rate', 'dst_host_count',
                   'dst_host_srv_count', 'dst_host_same_srv_rate', 'dst_host_diff_srv_rate', 'dst_host_serror_rate', 'dst_host_srv_serror_rate', 
                   'dst_host_rerror_rate']
```

```python
X_train = dataframe[features]
X_test = test_dataframe
labels = X_test['class']
results = pd.DataFrame(X_test, columns=dataframe.columns)
X_test = X_test[features]
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = IsolationForest(contamination=0.46)
model.fit(X_train)
```

```python
results['predictions'] = model.predict(X_test)
results['predictions'] = results['predictions'] == -1
results['predictions']
```

```python
results['is_anomaly'] = results['class'] == b'anomaly'
results['is_anomaly']
```

```python
true_anomalies = (results['predictions'] == True) & (results['is_anomaly'] == True)
sum(true_anomalies) / len(results['is_anomaly'])
```

```python
false_positives = (results['predictions'] == True) & (results['is_anomaly'] == False)
sum(false_positives) / len(results['is_anomaly'])
```

```python
false_negatives = (results['predictions'] == False) & (results['is_anomaly'] == True)
sum(false_negatives) / len(results['is_anomaly'])
```

```python
true_safe = (results['predictions'] == False) & (results['is_anomaly'] == False)
sum(true_safe) / len(results['is_anomaly'])
```

```python
precision = sum(true_anomalies) / (sum(true_anomalies) + sum(false_positives))
recall = sum(true_anomalies) / (sum(true_anomalies) + sum(false_negatives))
(precision, recall)
```

```python
f1 = 2 * precision * recall / (precision + recall)
f1
```
