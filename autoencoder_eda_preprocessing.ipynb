```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn 

```

```python
data_loc = "nsl-kdd/"
a1 = "KDDTrain+.txt"
a2 = "KDDTest+.txt"

import csv
file = open("nsl-kdd/labels.txt", "r")
data = list(csv.reader(file, delimiter=","))
data = data[0]

df_columns = pd.read_csv("nsl-kdd/labels.txt")
df = pd.read_csv(data_loc + a1, names=data, index_col=False)
df_test = pd.read_csv(data_loc + a2, names=data, index_col=False)
```

```python
traintest = pd.concat([df, df_test])
traintest
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(sparse=False)
enc = enc.fit(traintest[["flag"]])
ohe_columns = enc.get_feature_names_out()

ohe_df = pd.DataFrame(enc.transform(df[[ "flag"]]))
ohe_df_test = pd.DataFrame(enc.transform(df_test[["flag"]]))
ohe_df.columns = ohe_columns
ohe_df_test.columns = ohe_columns

```

```python
network_data_mod = ohe_df.join(df.drop(columns=["flag", "duration", "protocol_type", "service","src_bytes", "dst_bytes"]))
network_data_mod_test = ohe_df_test.join(df_test.drop(columns=["flag", "duration", "protocol_type", "service","src_bytes", "dst_bytes"]))
```

```python
ohe_df
```

```python
network_data_mod["class"] = np.where(network_data_mod["class"] == "normal", 1, 0)
network_data_mod_test["class"] = np.where(network_data_mod_test["class"] == "normal", 1, 0)
"""
network_data_mod['class'].mask(network_data_mod['class'] == 'normal', 1, inplace=True)
network_data_mod['class'].mask(network_data_mod['class'] == 'normal', 0, inplace=False)
network_data_mod_test['class'].mask(network_data_mod_test['class'] == 'normal', 1, inplace=True)
network_data_mod_test['class'].mask(network_data_mod_test['class'] == 'normal', 0, inplace=False)
"""
```
